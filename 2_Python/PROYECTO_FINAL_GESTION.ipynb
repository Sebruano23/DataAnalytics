## <font color='#71bebd'>**CONTEXTO**</font>
La base de datos de insumos de los equipos biomédicos es un registro electrónico que contiene información sobre los componentes y suministros necesarios para el funcionamiento y mantenimiento de los equipos médicos utilizados en la atención de la salud.

La base de datos de insumos es una herramienta valiosa para el personal encargado de la gestión de equipos biomédicos, ya que les permite realizar un seguimiento de los suministros utilizados en cada equipo, planificar y programar el mantenimiento preventivo, controlar los costos y evitar la interrupción del servicio debido a la falta de componentes o suministros necesarios. Además, también puede ser utilizada para llevar un control de las garantías y/o contratos de servicio de los equipos biomédicos.

Dentro de la ingeniería clinica hay diferente consideraciones para asegurar procesos óptimos en la gestión de tecnología:

* La **homogeneidad** de equipos biomédicos se refiere a la estandarización de los equipos y tecnologías utilizados en el ámbito médico para asegurar la consistencia y calidad en el rendimiento y resultados de los mismos. Esto implica una selección adecuada, evaluación y mantenimiento de los equipos biomédicos para garantizar su funcionamiento seguro y efectivo en el entorno clínico.

* La **tecnovigilancia** en equipos médicos se refiere al proceso de seguimiento y evaluación de los riesgos asociados con el uso de dispositivos médicos y tecnologías para prevenir y mitigar posibles problemas de seguridad y calidad en la atención al paciente. Esto incluye la identificación de riesgos potenciales, la recopilación de información y la implementación de medidas preventivas y correctivas para garantizar la seguridad del paciente y mejorar la calidad de la atención médica.

```python
equipos = inventario [["EQUIPO", "MARCA"]]
equipos = pd.DataFrame(equipos)
equipos["Cantidad"] = 1
equipos
```

```python
num_mar = equipos.groupby("MARCA").sum()
pd.DataFrame(num_mar)
```

```python
marcas_pred = num_mar[(num_mar["Cantidad"]>=20)]
marcas_pred.loc["OTROS"] = len(num_mar)-len(marcas_pred)
marcas_pred
```

```python
marcas = marcas_pred.index.tolist()
```

```python
sns.set_style("whitegrid")
my_palette = ["#003785", "#1465bb","#2196f3","#81c9fa","#b9ffff"]

# Crea un gráfico de torta con Seaborn
plt.figure(figsize=(10, 8))
sns.set_palette(my_palette)
plt.pie(marcas_pred["Cantidad"], labels=marcas, autopct='%1.1f%%')
plt.axis('equal')
plt.title("Lista de Marcas",fontdict={'fontsize': 18, 'fontweight': 'bold'})
plt.show()
```

Este análisis nos yuda a determinar que marcas de equipos han sido más usas, de esta manera se puede condicionar la elección de proveedores.

```python
prop_mar = num_mar["Cantidad"]/num_mar["Cantidad"].sum()
homogeneity_index = 1 - (1 / len(prop_mar)) * (prop_mar ** 2).sum()
print(homogeneity_index)
```
